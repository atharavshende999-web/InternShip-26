```python
# Write a program to find factorial using while loop
```

```python
fact=1
i=1
while i<=5:
    fact=fact*i
    i=i+1
print("Factorial of 5 is:-",fact)
    
```

```python
#Using User input
num=int(input("Enter the number"))
i=1
fact=1
while i<=num:
    fact=fact*i
    i=i+1
print(f"Factorial of {num} is {fact}")
```
