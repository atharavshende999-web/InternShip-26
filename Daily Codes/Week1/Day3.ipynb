```python
##Using Nested if
num=int(input("Enter Number"))
if(num>50):
    if(num<100):
        print("Number between 50 and 100")
    else:
        print("Number are Greater than 50 but not less than 100")
else:
    print("Not Number between 50 and 100")
```

# Menu Driven

```python
print("***Welcome to the Simple Calculator***")
num=int(input("Enter the Number"))
print("Menu =>\n1.Square\n2.Cube\n3.Check +ve and -ve")
ch=int(input("Enter your Choice"))

if ch==1:
    print("Square is:",num*num)
elif ch==2:
    print("Cube is:",num*num*num)
elif ch==3:
   
    if(num>0):
        print("Positive Number")
    elif num==0:
        print("Zero")
    else:
        print("Negative Number")
else:    
    print("Invlaid Choice")
    
```

```python
print("Calculator")
num1=int(input("Enter number1"))
num2=int(input("Enter number2"))
print("Menu=>\n1.Addition\n2.Subtraction\n3.Multiplication\n4.Divison")
ch=int(input("Enter your Choice"))
if ch==1:
    print("Addition is:",num1+num2)
elif ch==2:
    print("Subtraction is:",num1-num2)
elif ch==3:
    print("Multiplication is:",num1*num2)
elif ch==4:
    print("Division is:",num1/num2)
else:
    print("Invalid Choice")
    
```

```python

"""
menu driven->
taking a string from the user
menu -->upper,lowercase,replace a partiulat string
for third option it should which part to repalce with what

"""
```

```python
print("***Menu Driven***")
str=input("Enter String")
print("Menu\n1.LowerCase\n2.UpperCase\n3.Replace")
ch=int(input("Enter your choice"))
if ch==1:
    str=str.lower()
    print("LowerCase:",str)
elif ch==2:
    str=str.upper();
    print("UpperCase:",str)
elif ch==3:
    ch1=input("Enter the replace string")
    ch2=input("With what Letter")

    print("Replace String is",str.replace(ch1,ch2))
else:
    print("Invalid Choice")
```
