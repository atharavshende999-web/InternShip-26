# Inheritance

```python
class A:
    comp='Infosys'
    def __init__(self,age,name):
        self.age=age
        self.name=name
    def show(self):
        return f"Age is => {self.age},Name is=>{self.name},Company name is=>{self.comp}"
```

```python
class B(A):
    def __init__(self,age,name,sal):
        super().__init__(age,name)
        self.sal=sal
    def show1(self):
        print(f"Salary is=>{self.sal}")
```

```python
b=B(18,'ABC',123456)
```

```python
b.show1()
```

```python
b.show()
```

```python
class MenuApp:
    def __init__(self):
        self.listitem=[]
    def show_menu(self):
        print("Menu:\n1.Add Item\n2.View Item\n3.Exit")
    def add_item(self):
        add=input("Enter the item name")
        self.listitem.append(add)
    def view_item(self):
        print(self.listitem)
    def run(self):
        while True:
            self.show_menu()
            ch=int(input("Enter the Choice"))
            if ch==1:
                self.add_item()
            elif ch==2:
                self.view_item()
            elif ch==3:
                break
            else:
                print("Invalid Choice")
            
```

```python
m=MenuApp()
```

```python
m.run()
```

```python
class Person:
    def __init__(self,age,name):
        self.__age=age
        self.__name=name
    #getter
    def get_name(self):
        return self.__name
    def get_age(self):
        return self.__age
    #setter
    def set_age(self,age):
        self.__age=age
    def __str__(self):
        return f"age->{self.__age},name->{self.__name}"
    
```

```python
class MenuApp:
    def __init__(self):
        self.bal=5000
    
    def showmenu(self):
        print("***Menu:-***\n1.Check Balance\n2.Depoite\n3.WithDrawal\n4.Loan Elibile\n5.EMI Calculator\n6.Exit\n24*7 Customer Support")
    def balance(self):
        print(f"Current Balance is=>{self.bal}")
    def deposite(self):
        self.depo=int(input("Enter the amount to be Deposited=>"))
        self.bal+=self.depo
    def withdrawal(self):
        self.withdraw=int(input("Enter the Amount to be WithDrawal=>"))
        self.bal-=self.withdraw
    def loan_elig(self):
        self.cibil=int(input("Enter the Cibil Score"))
        if self.cibil>750 and self.cibil<=950:
            print("Eligible for Loan")
        elif self.cibil<700 and self.cibil>100:
            print("Not Eligible for loan")
        else:
            print("Please Enter Valid Cibile Score")
    def emi_cal(self):
        self.emi=0
        self.interest=float(input("Enter the Insterest"))
        self.prin=int(input("Enter the Principle"))
        self.year=int(input("Enter the Period of time"))
        r=self.interest/(12*100)
        n=self.year*12
        self.emi = (self.prin * r * (1 + r)**n) / ((1 + r)**n - 1)
        print(f"EMI is=>{round(self.emi,2)}")
    def run(self):
        while True:
            self.showmenu()
            ch=int(input("Enter your Choice"))
            if ch==1:
                self.balance()
            elif ch==2:
                self.deposite()
            elif ch==3:
                self.withdrawal()
            elif ch==4:
                self.loan_elig()
            elif ch==5:
                self.emi_cal()
            elif ch==6:
                break
            else:
                print("Enter Valid Choice")

        
```

```python
m=MenuApp()
```

```python
m.run()
```
