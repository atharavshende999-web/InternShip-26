```python
'''
take a string as user input
use a loop and if else to check if a specified character is present in the string or not
take 2 string to check a string
user input ->string,character1,character2
''';
```

```python
name=input("Enter the String")
ch1=input("Enter the Character1")
ch2=input("Enter the Character2")
flag1=0
flag2=0
for i in name:
    if ch1==i:
        flag1=1
        
    elif ch2==i:
        flag2=1

if flag1==1:
    print("Charater1 is Present")
if flag2==1:
    print("Charater2 is Present")
if flag1==0:
    print("Charater1 is Absent")
if flag2==0:
    print("Charater2 is Absent")
```
