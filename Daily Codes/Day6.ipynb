```python
l=[34,56,12,48,34,10]
```

```python
count=0
for i in l:
    print(f"{count} => {i}")
    count+=1
```

```python
for i in l:
    print(f"{l.index(i)} => {i}")
```

```python
list(enumerate(l)) ##It is iterate over list with combination of index=>value
```

```python
for ind,val in enumerate(l):
    print(f"{ind} => {val}")
```
