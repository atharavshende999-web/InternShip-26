```python
class Person:
    def __init__(atharv,name,age,height):
        atharv.name=name
        atharv.__age=age #Private
        atharv.height=height

    def info(self):
        return f"name:{self.name},Age:{self.__age},height:{self.height}"
    
```

```python
p1=Person('Atharv',17,55)
```

```python
p1.info()
```

```python
p1.age
```

```python
class Person:
    def __init__(atharv,name,age,height):
        atharv.name=name
        atharv._age=age #Protected
        atharv.height=height

    def info(self):
        return f"name:{self.name},Age:{self.age},height:{self.height}"
```

```python
Person.age=34
```

```python
Person.age
```

```python
Person.age
```

```python
class Empolyee:
    company="Wipro"
    def __init__(self,name,monthly_salary,job_designation):
        self.name=name
        self.__monthly_salary=monthly_salary
        self.job_designation=job_designation
    def info(self):
        return f"Company=>{self.company},Name=>{self.name},Monthly Salary=>{self.__monthly_salary}=>,Job Designation=>{self.job_designation}"
    def anual_salary(self):
        print("Anual Salary is=>",(self.__monthly_salary)*12)
    
        
```

```python
e1=Empolyee('ABC',15000,'HR')
e2=Empolyee('XYZ',18000,'Web Designer')
e3=Empolyee('PQR',10000,'Senior Engineer')
```

```python
e1.info()
```

```python
e2.info()
```

```python
e3.info()
```

```python
e1.anual_salary()
```

```python
e2.anual_salary()
```
