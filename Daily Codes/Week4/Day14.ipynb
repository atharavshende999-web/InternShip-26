```python
#We need to be specific about what kind of error we want to catch
```

```python
try:
    f=open('a1.txt')
    print(var1)
except FileNotFoundError:
    print("File Does Not Exist")
except Exception as e:
    print(e)
```

```python
##else->if try does not raise any exception
```

```python
try:
    f=open('a1.txt')
except FileNotFoundError:
    print("File Does Not Exist")
except Exception as e:
    print(e)
else:
    print(f.read())
    f.close()
```

```python
try:
    f=open('a1.txt')
    print(var1)
except FileNotFoundError:
    print("File Does Not Exist")
except Exception as e:
    print(e)
else:
    print(f.read())
    f.close()
finally:
    print("run completed")
```

```python
try:
    f=open('a1.txt')
    if f.name=='a1.txt':
        raise Exception
except FileNotFoundError:
    print("File Does Not Exist")
except Exception as e:
    print(e)
else:
    print(f.read())
    f.close()
finally:
    print("run completed")
```

```python
try:
    num1=int(input("Enter the Number1"))
    num2=int(input("Enter the Numer2"))
    div=num1//num2
    print(div)
except ZeroDivisionError:
    print(f"Can't Be Divided By Zero")
except ValueError:
    print(f"Invalid Datatype")
finally:
    print("run completed")
```
