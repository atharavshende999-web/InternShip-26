```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([10,23,29,34,39,45,50])
y=np.array([12,21,31,43,48,55,58])
z=np.array([56,50,44,37,26,20,16])
```

# Line Chart

```python
plt.plot(x,y);
```

```python
plt.xlabel("x value")
plt.ylabel("y value")
plt.title("Line Chart")
plt.plot(x,y,marker='o',mfc='red',ls='--');
plt.grid()
# o,*,H,<,>,^
```

```python
plt.plot(x,z,marker='*',mfc='green',ls='-')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Line Chart X vs Z")
plt.grid()
```
