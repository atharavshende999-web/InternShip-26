```python
'''
Create a Student class that stores marks for 5 subjects and has a method to calculate the percentage.
''';
```

```python
class Student:
    tot=0
    per=0
    def __init__(self,m1,m2,m3,m4,m5):
        self.m1=m1
        self.m2=m2
        self.m3=m3
        self.m4=m4
        self.m5=m5
    def calculate(self):
        self.tot=self.m1+self.m2+self.m3+self.m4+self.m5
        self.per=(self.tot/500)*100
        print("Percentage is=>",self.per)
        
```

```python
s=Student(89,89,89,67,99)
```

```python
s.calculate()
```
