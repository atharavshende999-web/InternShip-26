```python
list_a=[3,1,4]
list_b=[2,5,0]
list_combine=list_a+list_b
print(f"Combine List is:-{list_combine}")
```

```python

for i in range(len(list_combine)):   #6 0-5
    for j in range(0,len(list_combine)-1): #0,6-1
        if  list_combine[j]>list_combine[j+1]:
            list_combine[j],list_combine[j+1]=list_combine[j+1],list_combine[j]
print("After Swapping")    
print(list_combine)          
```

```python
nested_list=[1,[2,3],4,[5,6,7],8]
new_list=[]
```

```python
for i in nested_list:
    if isinstance(i,list):
        for j in i:
            new_list.append(j)
    else:
        new_list.append(i)
print("Flat List=>",new_list)
```
