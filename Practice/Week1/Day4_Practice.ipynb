```python
## Water Tank=5000l
```

```python
tank_capacity=5000
tank_free_capacity=0
water_added=0
while True:
    add=int(input("Enter the Water to be added"))
    water_added+=add
    if water_added>=5000:
        print("Warning:Tank Overflow")
        break
    else:
        print(f"Water Added {water_added} in Tank")
        tank_free_capacity=tank_capacity-water_added
        print("Tank Free Capcity is:",tank_free_capacity)
print("***Save Water***")
    
```
