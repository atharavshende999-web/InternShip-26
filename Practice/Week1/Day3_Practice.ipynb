```python
num=int(input("Enter the number"))
flag=0
if(num<0):
    print("Please enter the positive number")
elif num>0:
    for i in range(2,num-1):
        if num%i==0:
            flag=1
            break
if(flag==1):
    print("Not Prime Number")
else:
    print("Prime Numer")
            
```
