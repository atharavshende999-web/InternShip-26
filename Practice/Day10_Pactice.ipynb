```python
class A:
    def show(self):
        print("A")

class B(A):
    def show(self):
        super().show()
        print("B")

class C(B):
    def show(self):
        super().show()
        print("C")

obj = C()
obj.show()
```

```python
class Shop:
    def __init__(self):
        self.tot = 0

    def item(self):
        print("\n--- Shopping Cart System ---")
        print("1. Electronic\n2. Clothing\n3. Exit")
        ch = int(input("Enter Choice: "))

        if ch == 1:
            print("\nElectronic Category")
            print("1. Laptop\n2. Mobile\n3. Smart Watch\n4. TWS")
            ch1 = int(input("Enter Product: "))
            input("Press Enter to view details...")

            if ch1 == 1:
                print("""Name: Laptop
Brand: Lenovo
RAM: 8 GB
Processor: Intel Core i5 (11th Gen)
Price: ₹55,000""")
                input("Press Enter to add to cart...")
                self.tot += 55000

            elif ch1 == 2:
                print("""Name: Redmi Note 13
Brand: Xiaomi
RAM: 6 GB
Processor: Snapdragon 685
Price: ₹12,000""")
                input("Press Enter to add to cart...")
                self.tot += 12000

            elif ch1 == 3:
                print("""Name: Noise ColorFit Pro 5
Brand: Noise
Price: ₹3,000""")
                input("Press Enter to add to cart...")
                self.tot += 3000

            elif ch1 == 4:
                print("""Name: boAt Airdopes 141
Brand: boAt
Price: ₹1,200""")
                input("Press Enter to add to cart...")
                self.tot += 1200

            else:
                print("Invalid Choice")

        elif ch == 2:
            print("\nClothing Category")
            print("1. Men\n2. Women\n3. Kids")
            ch1 = int(input("Enter Product: "))
            input("Press Enter to view details...")

            if ch1 == 1:
                print("""Name: Casual T-Shirt
Brand: Levi’s
Price: ₹1,500""")
                input("Press Enter to add to cart...")
                self.tot += 1500

            elif ch1 == 2:
                print("""Name: Kurti
Brand: Biba
Price: ₹2,500""")
                input("Press Enter to add to cart...")
                self.tot += 2500

            elif ch1 == 3:
                print("""Name: Kids Wear Set
Brand: Babyhug
Price: ₹1,200""")
                input("Press Enter to add to cart...")
                self.tot += 1200

            else:
                print("Invalid Choice")

        elif ch == 3:
            return False

        else:
            print("Invalid Choice")

        return True

    def total(self):
        print(f"\nTotal Amount to Pay: ₹{self.tot}")


s = Shop()

while True:
    if not s.item():
        break

s.total()
```
