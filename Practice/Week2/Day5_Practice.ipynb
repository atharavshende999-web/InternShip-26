```python
l4=[1,2,3,4,1,2,3,4,4]
rep=[]
```

```python
for i in l4:
    if l4.count(i)>1 and i not in rep:
        rep.append(i)
print(rep)

            
```

```python
l4=[1,2,3,4,1,2,3,4,4]


for i in l4:
    max=l4[0]
    min=l4[0]
    if max>i:
        max=i
    else:
        min=i
print(max)
print(min)
```
