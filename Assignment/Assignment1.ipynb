```python
## Accepting input from user until it give 0  and calculate sum and average and does'nt include 0 in calculation
```

```python
count=0
sum=0
avg=0
while True:
    n=int(input("Enter the Number"))
    if n==0:
        break
    else:
        count+=1
        sum+=n
avg=sum//count
print("Addition is:",sum)
print("Average is:",avg)
        
```
