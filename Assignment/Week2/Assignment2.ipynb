```python
def inventory_manager(record):
    result={}
    for product,category,quantity,price in record:
        if category not in result:
            result[category]={
                'total_items':0,
                'total_value':0,
                'product':[]
            }
        result[category]['total_items']+=quantity
        result[category]['total_value']+=quantity*price
        result[category]['product'].append(product)
    return result
    
    
```

```python
record=[("Laptop","Electronic",10,55000),("Mouse","Electronic",50,499),("Desk","Furniture",8,12000),("Chair","Furniture",15,4500),("Pen","Stationary",20,200)]
d=inventory_manager(record)
print(d)
```
