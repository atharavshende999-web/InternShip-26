```python
class BankAccount:
    balance=0
    def __init__(self):
        self.account_holder=input("Enter the Account Holder")
        self.account_number=int(input("Enter the Account Number"))
        self.balance=float(input("Enter the Current Balance"))
    def deposite(self,amount):
        self.balance=self.balance+amount
        print(f"Deposite Successfully!")
    def withdraw(self,amount):
        if self.amount>self.balance:
            pass
        else:
            self.balance-=amount
            print(f"Withdrawal Successfully!")
        
    def get_balance(self):
        print("Balance is=>",self.balance)
    def __str__(self):
        return f"Account Holder is->{self.account_holder},Amount Number is->{self.account_number},Balance is->{self.balance}"
```

```python
import math as m
class Shape:
    def area(self):
        return 0;
    def describe(self):
        print("I am a shape")
class Circle(Shape):
    def area(self,radius):
        print(f"Area of Circle is ->{m.pi*radius*radius}")
    def describe(self):
        print("I am circle")
class Rectangle(Shape):
    def area(self,length,width):
        print(f"Area of Rectangle ->{length*width}")
    def describe(self):
        print("I am Rectangle")
class Triangle(Shape):
    def area(self,base,height):
        print(f"Area of Triangle is ->{base*height}")
    def describe(self):
        print(f"I am Triangle")
        
    
```

```python
class Vehicle:
    def __init__(self,make,model,year):
        self.make=make
        self.model=model
        self.year=year
    def start_engine(self):
        print("Engine Started")
class Car(Vehicle):
    def __init__(self,make,model,year,num_door):
        super().__init__(make,model,year)
        self.num_door=num_door
    def start_engine(self):
        print(f"{self.model} is started")
class ElectriCar(Vehicle):
    def __init__(self,make,model,year,battery_capacity):
        super().__init__(make,model,year)
        self.battery_capacity=battery_capacity
    def start_engine(self):
        print(f"Silent Start at 10 kWh")
```
