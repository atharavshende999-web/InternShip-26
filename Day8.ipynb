```python
#Function to check whether the number is greater of two number
```

```python
def greater(num1,num2):
    if num1>num2:
        print(f"{num1} is greater")
    else:
        print(f"{num2}is greater")
```

```python
greater(12,23)
```

```python

def number(num):
    fact=1
    i=1
    while i<=num:
        fact=fact*i
        i=i+1
    print(fact)
    
```

```python
number(5)
```
